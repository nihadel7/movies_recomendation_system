```python
import pandas as pd
movies_df=pd.read_csv('data1.csv')
```

```python

# Initialize an empty list to store user-item interactions
interactions_data = []

# Iterate through the original dataset and extract user-item interactions
for index, row in movies_df.iterrows():
    user_ids = row['user_ids']
    movie_title = row['movie_title']
    rating = row['rating']
    interactions_data.append({'user_id': user_ids, 'movie_title': movie_title, 'rating': rating})

# Create a new DataFrame from the extracted interactions data
user_interactions_df = pd.DataFrame(interactions_data)

# Display the new DataFrame
user_interactions_df.head()
```

```python
import pandas as pd

# Load your dataset containing user interactions (user_id, movie_title, rating)
  # Replace "your_dataset.csv" with the filename of your dataset

# Split the string of user IDs into separate rows
user_interactions_df['user_id'] = user_interactions_df['user_id'].str.strip("[]").str.split(", ")

# Explode the 'user_id' column to have each user on its own row
user_interactions_df = user_interactions_df.explode('user_id')

# Example Usage
user_interactions_df.head(20)
```

```python
import pandas as pd

# Define a function for movie-based collaborative filtering recommendations
def movie_based_recommendations(movie_title, user_interactions_df, n=10):
    # Filter interactions for the given movie title
    movie_interactions = user_interactions_df[user_interactions_df['movie_title'] == movie_title]

    # Get users who rated the given movie highly (e.g., rating >= 4)
    high_rated_users = movie_interactions[movie_interactions['rating'] >= 4]['user_id'].unique()

    # Get interactions of high-rated users
    high_rated_users_interactions = user_interactions_df[user_interactions_df['user_id'].isin(high_rated_users)]

    # Group by movie_title and aggregate ratings
    aggregated_ratings = high_rated_users_interactions.groupby('movie_title')['rating'].mean()

    # Sort aggregated ratings to get recommendations
    recommendations = aggregated_ratings.sort_values(ascending=False).head(n)

    return recommendations

# Example Usage
movie_title = input("Enter the movie title: ")
recommendations = movie_based_recommendations(movie_title, user_interactions_df)
print("Recommendations based on", movie_title, ":")
for movie, rating in recommendations.items():
    print(movie, "-", "Rating:", rating)
```
